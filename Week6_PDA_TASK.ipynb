```python
import pandas as pd

# Load the dataset
df = pd.read_excel('/content/Global Superstore Orders 2016.xlsx')

# Display the number of rows and columns
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

# List all column names
print("\nColumn names:")
for col in df.columns:
    print(col)

# Show the first five rows
print("\nFirst five rows of the dataset:")
display(df.head())

# Display summary statistics for numeric columns
print("\nSummary Statistics:")
display(df.describe())
```

```python
df['Profit_Per_Sale'] = df['Profit'] / df['Sales']
print("DataFrame with new 'Profit_Per_Sale' column:")
display(df.head())
```

The `Profit_Per_Sale` column indicates the profit margin for each sale. A higher value means a larger profit generated per unit of sales revenue. It helps to understand the efficiency of each sale in terms of profitability.

```python
highest_profit_ratio_order = df.loc[df['Profit_Per_Sale'].idxmax()]
print("\nOrder with the highest profit ratio:")
display(highest_profit_ratio_order)
```

The order with the highest profit ratio (Profit_Per_Sale) is displayed above. This indicates the single transaction that yielded the highest profit margin relative to its sales value.

```python
df.to_excel("Cleaned_Global_Superstore.xlsx", index=False)
print("The cleaned data has been saved to 'Cleaned_Global_Superstore.xlsx'")
```

The saved file is named **`Cleaned_Global_Superstore.xlsx`**.

Exporting data after cleaning and analysis is useful for several reasons:

1.  **Preservation of Work**: It saves all the transformations, new features (like `Profit_Per_Sale`), and cleaning steps applied to the original data. This means you don't have to rerun all the cleaning and processing steps every time you want to use the data.
2.  **Shareability**: The cleaned dataset can be easily shared with others who might need to use it for further analysis, reporting, or building models, without needing access to the original raw data or the processing scripts.
3.  **Foundation for Further Analysis**: It provides a clean, ready-to-use dataset for subsequent stages of a project, such as advanced modeling, dashboard creation, or specific business intelligence tasks.
4.  **Version Control/Reproducibility**: Saving intermediate versions of the data can act as a form of version control, making it easier to revert to a clean state or reproduce results from a specific point in the analysis.
5.  **Performance**: Reading a pre-cleaned file is often faster than loading raw data and then applying all cleaning and preprocessing steps each time.

```python
# Group by Category and calculate total Profit
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
print("\nTotal Profit by Category:")
display(profit_by_category.sort_values(by='Profit', ascending=False))
```

```python
# Group by Region and calculate total Profit
profit_by_region = df.groupby('Region')['Profit'].sum().reset_index()
print("\nTotal Profit by Region:")
display(profit_by_region.sort_values(by='Profit', ascending=False))
```

```python
# Sort orders by Profit from highest to lowest
df_sorted_profit_desc = df.sort_values(by='Profit', ascending=False)

print("\nOrders sorted by Profit (highest to lowest) - Top 5 rows:")
display(df_sorted_profit_desc.head())
```

```python
# Filter rows where Sales are greater than 500
sales_gt_500 = df[df['Sales'] > 500]
print(f"Number of orders with Sales > 500: {len(sales_gt_500)}")
print("\nFirst 5 rows with Sales > 500:")
display(sales_gt_500.head())
```

```python
# Filter rows where Sales are greater than 1000
sales_gt_1000 = df[df['Sales'] > 1000]
print(f"Number of orders with Sales > 1000: {len(sales_gt_1000)}")
print("\nFirst 5 rows with Sales > 1000:")
display(sales_gt_1000.head())
```
