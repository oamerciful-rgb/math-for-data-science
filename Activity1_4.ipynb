```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, None, 35, None],
    'email': ['a@email.com', None, 'c@email.com', 'd@email.com'],
    'signup_data': ['2023-01-10', '2023-03-20', '2023-03-25', None] # Added None to match length
}

df = pd.DataFrame(data)

# task 1 - identify missing values and incorrect data types
print(df.isnull().sum())
# shown how many NaN per column
# signup_data will show 'object' instead of datetime

# task 2 - fill missing ages with median AGE
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

# task 3 - remove rows with missing email addresses
df = df.dropna(subset=['email'])

# task 4 - convert signup_data to datetime
df['signup_data'] = pd.to_datetime(df['signup_data'])

print(df)
print(df.dtypes)
```
