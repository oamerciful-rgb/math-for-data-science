```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Step 1: Data
x = np.linspace (-5, 5, 40)
y = 3*x**2 + 2*x + 5 + 2*np.random.randn(len(x)) # Added noise to y

# Step 2: Model
def model (x, a, b, c):
    return a*x**2 + b*x + c

params, _ = curve_fit (model, x, y)

# Step 3: Turning Point
a, b, c = params # Unpack parameters
x_min = -b / (2*a)
y_min = model (x_min, a, b, c)

# Step 4: Plot
plt.scatter (x, y)
plt.plot (x, model (x, *params), color='r')
plt.scatter (x_min, y_min, color='r', s=80)
plt.show ()
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
x =  iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

df = pd.DataFrame(x, columns=feature_names)
df['target'] = y
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
dates = pd.date_range(start="2024-01-01", periods=120)

traffic = 100 + np.cumsum(np.random.normal(0.5, 2, size=len(dates)))
conversion = 2 + np.random.normal(0, 0.2, size=len(dates))


event_date = pd.Timestamp("2024-03-15")
event_idx = dates.get_loc(event_date)

traffic[event_idx:] += np.linspace(0, 80, len(dates) - event_idx)
conversion[event_idx:] += np.linspace(0, 1.5, len(dates) - event_idx)

df = pd.DataFrame({
    "date": dates,
    "traffic": traffic,
    "conversion": conversion
}).set_index("date")

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)


ax1.plot(df.index, df["traffic"], color="tab:blue", linewidth=2.5, label="Traffic")
ax1.fill_between(df.index, df["traffic"], color="tab:blue", alpha=0.1)

ax1.set_title("Website Growth Accelerates After Campaign Launch", fontsize=14, weight='bold')
ax1.set_ylabel("Traffic", fontsize=11)

ax1.axvline(event_date, color="red", linestyle="--", alpha=0.7)
ax1.annotate(
    "Campaign Launch",
    xy=(event_date, df.loc[event_date, "traffic"]),
    xytext=(event_date, df["traffic"].max() * 0.7),
    arrowprops=dict(facecolor='red', shrink=0.05),
    fontsize=10,
    color="red"
)



ax2.plot(df.index, df["conversion"], color="tab:green", linewidth=2.5, label="Conversion Rate")


ax2.axvspan(event_date, df.index[-1], color="green", alpha=0.08)

ax2.set_ylabel("Conversion Rate (%)", fontsize=11)
ax2.set_xlabel("Date", fontsize=11)

ax2.annotate(
    "Sustained improvement\nin conversions",
    xy=(df.index[-20], df["conversion"].iloc[-20]),
    xytext=(df.index[-60], df["conversion"].max() * 0.9),
    arrowprops=dict(facecolor='black', arrowstyle="->"),
    fontsize=10
)

for ax in (ax1, ax2):
    ax.spines[['top', 'right']].set_visible(False)
    ax.grid(alpha=0.2)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ── Load dataset ──────────────────────────────────────────────────
# Upload the file using the sidebar folder icon, then run this cell
df = pd.read_excel('/content/Global Superstore Orders 2016.xlsx')

print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
df[['Category', 'Region', 'Sales', 'Profit', 'Country', 'Ship Mode']].head(3)
```

```python
# Group by Category and sum Sales
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(8, 5))

bars = ax.bar(cat_sales.index, cat_sales.values,
              color=['#2196F3', '#FF9800', '#4CAF50'], edgecolor='white', linewidth=0.8)

# Add value labels on bars
for bar in bars:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 20000,
            f'${bar.get_height():,.0f}',
            ha='center', va='bottom', fontsize=10, fontweight='bold')

ax.set_title('Total Sales by Product Category — Global Superstore 2016',
             fontsize=13, fontweight='bold', pad=12)
ax.set_xlabel('Product Category', fontsize=11)
ax.set_ylabel('Total Sales (USD)', fontsize=11)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()

print('\nReflection: Which category brings the highest total sales?')
print(f'→ {cat_sales.idxmax()} with ${cat_sales.max():,.2f} in total sales.')
```
